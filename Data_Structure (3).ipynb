```python
#1. Discuss string slicing and provide examples.
#1. String slicing is a powerful feature in Python that allows you to extract a subset of characters from a string. It's done by using square brackets [] with a colon : to specify the start and end indices of the slice.

#Basic syntax:

#string[start:end]

#Where:

#- start is the index where the slice starts (inclusive)
#- end is the index where the slice ends (exclusive)

#Examples:

#1. Get the first 3 characters:

#my_string = "Hello World"
#print(my_string[0:3])  # Output: "Hel"

#1. Get the last 3 characters:

#print(my_string[-3:])  # Output: "rld"

#1. Get the substring from index 4 to 8:

#print(my_string[4:8])  # Output: "o Wo"

#1. Get the entire string:

#print(my_string[:])  # Output: "Hello World"

#1. Get the string from index 3 to the end:

#print(my_string[3:])  # Output: "lo World"

#1. Get the string from the beginning to index 5:

#print(my_string[:5])  # Output: "Hello"

#Note:

#- Negative indices count from the end of the string.
#- If start is omitted, it defaults to 0.
#- If end is omitted, it defaults to the end of the string.
#- If start is greater than end, an empty string is returned.

#String slicing is a versatile tool for text manipulation and extraction in Python.
```

```python
#2. Explain the key features of lists in Python.

#2. Lists in Python are a fundamental data structure that offers a wide range of features, making them a versatile tool for storing and manipulating data. Here are the key features of lists:

#1. Ordered collection: Lists maintain the order in which elements are added or inserted.

#2. Mutable: Lists can be modified after creation (e.g., elements can be added, removed, or changed).

#3. Indexing: Lists support indexing, allowing access to individual elements by their index (position).

#4. Slicing: Lists can be sliced to extract subsets of elements.

#5. Dynamic size: Lists can grow or shrink dynamically as elements are added or removed.

#6. Heterogeneous: Lists can store elements of different data types (e.g., numbers, strings, objects).

#7. Nested: Lists can contain other lists (nested lists).

#8. Methods: Lists have various built-in methods for manipulation, such as append(), insert(), remove(), sort(), and more.

#9. Concatenation: Lists can be concatenated using the + operator.

#10. Iteration: Lists can be iterated over using loops (e.g., for loops).

#These features make lists a powerful and flexible data structure in Python, suitable for a wide range of applications, from simple data storage to complex data processing and manipulation.
```

```python
#3. Describe how to access, modify, and delete elements in a list with examples.
#3. Here's how to access, modify, and delete elements in a list:

#Accessing Elements:

#- Use indexing to access a single element: list[index]
#- Use negative indexing to access elements from the end: list[-index]
#- Use slicing to access a subset of elements: list[start:end]

#Example:

#my_list = [1, 2, 3, 4, 5]
#print(my_list[0])  # Output: 1
#print(my_list[-1])  # Output: 5
#print(my_list[1:3])  # Output: [2, 3]

#Modifying Elements:

#- Assign a new value to an existing index: list[index] = new_value
#- Use the append() method to add an element to the end: list.append(new_value)
#- Use the insert() method to add an element at a specific index: list.insert(index, new_value)

#Example:

#my_list = [1, 2, 3, 4, 5]
#my_list[0] = 10
#print(my_list)  # Output: [10, 2, 3, 4, 5]
#my_list.append(6)
#print(my_list)  # Output: [10, 2, 3, 4, 5, 6]
#my_list.insert(2, 7)
#print(my_list)  # Output: [10, 2, 7, 3, 4, 5, 6]

#Deleting Elements:

#- Use the del statement to delete an element at a specific index: del list[index]
#- Use the remove() method to delete the first occurrence of a value: list.remove(value)
#- Use the pop() method to delete an element at a specific index and return its value: list.pop(index)

#Example:

#my_list = [1, 2, 3, 4, 5]
#del my_list[0]
#print(my_list)  # Output: [2, 3, 4, 5]
#my_list.remove(3)
#print(my_list)  # Output: [2, 4, 5]
#print(my_list.pop(0))  # Output: 2
#print(my_list)  # Output: [4, 5]

#These are the basic ways to access, modify, and delete elements in a list.
```

```python
#4. Compare and contrast tuples and lists with examples.
#4. Tuples and lists are both data structures in Python used to store collections of values. Here's a comparison of their key features:

#Similarities:

#1. Both can store multiple values.
#2. Both support indexing and slicing.
#3. Both can be nested.

#Differences:

#1. Immutability: Tuples are immutable, meaning their values cannot be changed after creation. Lists are mutable, meaning their values can be changed.
#2. Syntax: Tuples use parentheses () while lists use square brackets [].
#3. Performance: Tuples are faster and more memory-efficient than lists.

#Examples:

#Tuple:

#my_tuple = (1, 2, 3, 4, 5)

#- Immutable:
    #- my_tuple[0] = 10 raises a TypeError
#- Indexing and slicing:
    #- print(my_tuple[0]) outputs 1
    #- print(my_tuple[1:3]) outputs (2, 3)

#List:

#my_list = [1, 2, 3, 4, 5]

#- Mutable:
    #- my_list[0] = 10 changes the first element to 10
#- Indexing and slicing:
    #- print(my_list[0]) outputs 1
    #- print(my_list[1:3]) outputs [2, 3]

#When to use each:

#- Use tuples for:
    #- Storing constant data
    #- Returning multiple values from a function
    #- Using as dictionary keys
#- Use lists for:
    #- Storing dynamic data
    #- Modifying or updating data
    #- Inserting or removing elements

#In summary, tuples are immutable, faster, and more memory-efficient, while lists are mutable and more flexible. Choose the right data structure based on your use case!
```

```python
#8. Discuss the importance of dictionary keys being immutable and provide examples.
#8. In Python, dictionary keys must be immutable, meaning their value cannot change after creation. This is crucial for several reasons:

#1. Hashing: Dictionary keys are hashed to store and retrieve values efficiently. If keys were mutable, their hash value would change, making it impossible to find the associated value.

#2. Uniqueness: Immutable keys ensure that each key is unique. If keys could change, duplicates might occur, violating the dictionary's fundamental property.

#3. Performance: Immutable keys enable fast lookups, as the hash value is computed only once. If keys were mutable, the hash would need to be recomputed every time the key changes.

#Examples of immutable types suitable for dictionary keys:

#- Integers (e.g., 1, 2)
#- Floats (e.g., 3.14)
#- Strings (e.g., "hello", "world")
#- Tuples (e.g., (1, 2), ("a", "b"))

#Examples of mutable types that cannot be used as dictionary keys:

#- Lists (e.g., [1, 2], ["a", "b"])
#- Dictionaries (e.g., {"a": 1}, {"b": 2})
#- Sets (e.g., {1, 2}, {"a", "b"})

#If you try to use a mutable type as a dictionary key, Python raises a TypeError:

#my_dict = {[1, 2]: "value"}  # Raises TypeError: unhashable type: 'list'

#In summary, the immutability of dictionary keys ensures efficient storage, uniqueness, and fast lookups, making dictionaries a powerful data structure in Python.
```
